```python
import joblib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_auc_score
import numpy as np
```

```python
results = joblib.load('../4_evaluation/large/DRH/results.sav')['real']

scores = []
for key in results.keys():
    print(key)
    labels = key.split(' vs. ')
    model = results[key]['model']
    X_test = results[key]['X_test']
    y_test = results[key]['y_test']
    display(pd.concat([y_test, X_test], axis=1).groupby('Outcome').quantile(0.75))
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    score = roc_auc_score(y_true=y_test, y_score=y_proba)
    scores.append(score)
    print('ROC-AUC:', score)

np.mean(scores)
```
